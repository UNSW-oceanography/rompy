```python
import rompy
cat = rompy.cat
df=cat.aodn.nrt_wave_altimetry.read()
test_url_1="http://thredds.aodn.org.au/thredds/dodsC/"+df.file_url[0]
test_url_2="http://thredds.aodn.org.au/thredds/dodsC/"+df.file_url[1]
import xarray as xr
ds_1=xr.open_dataset(test_url_1)
ds_2=xr.open_dataset(test_url_2)
```

```python
ds = xr.concat([ds_1,ds_2], dim='TIME', 
                coords=['TIME'], 
                compat="override", 
                combine_attrs="override")
```
